```python
# Given a string s, find the length of the longest substring without repeating characters.

 

# Example 1:

# Input: s = "abcabcbb"
# Output: 3
# Explanation: The answer is "abc", with the length of 3.
# Example 2:

# Input: s = "bbbbb"
# Output: 1
# Explanation: The answer is "b", with the length of 1
def lengthOfLongestSubstring(self, s: str) -> int:
        w = []
        word = ""
        if(len(s) == 0) :
            return 0
        elif (s == " ") :
            return 1
        elif len(s) == 1 :
            return 1
            
        for l in range(len(s)) :
            word = s[l]
            if(word == " ") :
                w.append(word)
                word = ""
                continue
            #print(word)
            for l2 in range(l+1,len(s)) :
                if(s[l2] in word) :
                    #print("word",s[l2])
                    break
                else :
                    word = word + s[l2]
                    #print(word)
            #print(word)        
            if(len(word) > 0 ) :
                w.append(word)
                
                word = ""
        
        len1 = 0
        for i in w :
            
            #print(w)
            len1 = max(len(i),len1)
        return len1
```

```python
# Given an m x n integer matrix matrix, if an element is 0, set its entire row and column to 0's.

# You must do it in place.

 

# Example 1:


# Input: matrix = [[1,1,1],[1,0,1],[1,1,1]]
# Output: [[1,0,1],[0,0,0],[1,0,1]]
    
def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        row = len(matrix)
        col = len(matrix[0])
        l = []
        def setzero(matrix,i,j)   :
            for r in range(col) :
                matrix[i][r] = 0
            for c in range(row) :
                matrix[c][j] = 0
            return
        
        for i in range(row):
            for j in range(col) :
                if(matrix[i][j] == 0) :
                    l.append([i,j])
                    
        for  i in range(len(l))   :
            setzero(matrix,l[i][0],l[i][1])
       
```

```python
def max_sub(A, B):
    m, n = len(A), len(B)
    dp = [[0 for _ in range(m+1)] for _ in range(n+1)]

    mx = 0
    for i in range(1, n+1):
        for j in range(1, m+1):
            if A[i-1] == B[j-1]:
                dp[i][j] = 1 + dp[i-1][j-1]
                mx = max(mx, dp[i][j])
    return mx
    

useridList =  ["user1", "user2", "user3", "user4"]
categoryList = [["fashion","restaurant","grocery","utility"], ["fuel","fashion","restaurant","grocery"], ["fuel","utility","restaurant","grocery"], ["internet","transportation","water","grocery"]]

mx = 0
res = {}
for i in range(len(useridList)):
    for j in range(i + 1, len(useridList)):
        count = max_sub(categoryList[i], categoryList[j])
        res[count] = [useridList[i], useridList[j]]
        mx = max(mx, count)

print( res[mx])
```

```python
# Given an m x n grid of characters board and a string word, return true if word exists in the grid.
# The word can be constructed from letters of sequentially adjacent cells, where adjacent cells are horizontally or vertically neighboring. The same letter cell may not be used more than once.
# Example 1:
# Input: board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]], word = "ABCCED"
# Output: true

class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        r = len(board)
        c = len(board[0])
        currentrow = 0
        currentcol = 0
        
        def search(i,j,let) :
            for i in range(r) :
                #print(i,i)
                for j in range(c) :
                    #print(j,j)
                    if(board[i][j] == let) :
                        board[i][j] ="V" + let
                        return (i,j,True)
            return(i,j,False)
        for let in word :
            currentrow,currentcol,status = search(currentrow,currentcol,let)
            #print(currentrow,currentcol,status)
            if(status == False) :
                return False
            
        return status         
        
        
```

```python
# Given an integer array nums, return true if there exists a triple of indices (i, j, k) such that i < j < k and nums[i] < nums[j] < nums[k]. If no such indices exists, return false.
# Example 1:
# Input: nums = [1,2,3,4,5]
# Output: true
# Explanation: Any triplet where i < j < k is valid.
    def increasingTriplet(self, nums: List[int]) -> bool:
        
        first_num = float("inf")
        second_num = float("inf")
        print(first_num,second_num)
        for n in nums:
            if n <= first_num:
                first_num = n
            elif n <= second_num:
                second_num = n
            else:
                return True
        return False
        
```

```python
# You are given an integer array coins representing coins of different denominations and an integer amount representing a total amount of money.

# Return the fewest number of coins that you need to make up that amount. If that amount of money cannot be made up by any combination of the coins, return -1.

# You may assume that you have an infinite number of each kind of coin.

 

# Example 1:

# Input: coins = [6,2,5], amount = 11
# Output: 3
#Explanation: 11 = 5 + 5 + 1
def coinChange(coins, amount ) -> int:
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    print(dp)

    for coin in coins:
        for x in range(coin, amount + 1):
            dp[x] = min(dp[x], dp[x - coin] + 1)
            print(dp[x],dp[x - coin] + 1,x,x - coin)
    return dp[amount] if dp[amount] != float('inf') else -1 
print(coinChange([6,2,5],11))

```

```python
# Given an array of strings strs, group the anagrams together. You can return the answer in any order.

# An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

 

# Example 1:

# Input: strs = ["eat","tea","tan","ate","nat","bat"]
# Output: [["bat"],["nat","tan"],["ate","eat","tea"]]
import collections
def groupAnagrams(strs) :
    ans = collections.defaultdict(list)
    for s in strs:
        print(sorted(s))
        #print(tuple(sorted(s)))
        print("hhh")
        ans[tuple(sorted(s))].append(s)
    return ans.values()
    
l = ["eat","tea","tan","ate","nat","bat"]
groupAnagrams(l)
```
